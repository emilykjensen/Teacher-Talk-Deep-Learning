```python
import pandas as pd
```

```python
df = pd.read_csv('utterance-data-with-labels-binary.csv')
df.head()
df.shape
```

```python
for i in df:
    print(i)
```

```python
df = df.drop(columns=['IsInstructionalStatement','IsDisciplinaryUtterance','IsDisciplinaryStatement','IsEvaluationFollowupIncluded','IsEvaluationValencePositive','CombinedAuthCogUptake','Uptake','IsInstructionalQuestion','IsDisciplinaryQuestion','IsStudentResponsePresent','IsSerialQuestion'])
```

```python
df.head()
```

```python
for i in df:
    print(i)
```

```python
df['labels'] = list(zip(df.IsQuestion.tolist(), df.IsInstructionalUtterance.tolist(), df.IsEvaluationElaborated.tolist(), df.CogLevel.tolist(), df.IsGoalSpecified.tolist(), df.IsDisciplinaryTermsPresent.tolist(), df.Authenticity.tolist()))
df['text'] = df['TranscribedUtterance'].apply(lambda x: x.replace('\n', ' '))

df.head()
```

```python
#df.tail()
```

```python
from sklearn.model_selection import train_test_split
train_df, eval_df = train_test_split(df, test_size=0.2)
```

```python
#This cell trains the model. Since, I already trained the model we don'y have to execute it.

# from simpletransformers.classification import MultiLabelClassificationModel
# model = MultiLabelClassificationModel('roberta', 'roberta-base', num_labels=18, args={'train_batch_size':2, 'gradient_accumulation_steps':16, 'learning_rate': 3e-5, 'num_train_epochs': 3, 'max_seq_length': 512})

#model.train_model(train_df)
```

```python
# #To load a model a previously saved model instead of a default model, you can change the model_name to the path to a directory which contains a saved model.


from simpletransformers.classification import MultiLabelClassificationModel
model = MultiLabelClassificationModel('roberta', 'outputs/', num_labels=7, use_cuda=False)
```

```python
model.model.roberta
```

```python
model.model.get_output_embeddings
```

```python
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```

```python
#This is the evaluation results of the model
print(result)
```

```python
#The above evaluation command does not give us the predicted output so we run the following cells to get the predicted value and perform various types of performance tests.
#This is the test dataframe to be predicted
eval_df.head()
```

```python
to_predict = eval_df.text.apply(lambda x: x.replace('\n', ' ')).tolist()
preds, outputs = model.predict(to_predict)

#sub_df = pd.DataFrame(outputs, columns=['toxic','severe_toxic','obscene','threat','insult','identity_hate'])
sub_df = pd.DataFrame(outputs, columns=['IsQuestion','IsInstructionalUtterance','IsEvaluationElaborated','CogLevel','IsGoalSpecified','IsDisciplinaryTermsPresent','Authenticity'])

sub_df['ObsID'] = eval_df['ObsID']
sub_df = sub_df[['ObsID', 'IsQuestion','IsInstructionalUtterance','IsEvaluationElaborated','CogLevel','IsGoalSpecified','IsDisciplinaryTermsPresent','Authenticity']]

sub_df.to_csv('outputs/submission_our_project.csv', index=False)
```
